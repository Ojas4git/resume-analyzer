```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#step2 load dataset
data = pd.read_csv("jobs.csv")
print(data.head())
```

```python
#step3 resume input
resume = "python machine learning data analysis"
```

```python
#step4 combine data
texts = data["skills"].tolist()
texts.append(resume)
```

```python
#step5 Tfidf Vectorization
vectorizer = TfidfVectorizer()
vectors = vectorizer.fit_transform(texts)
```

```python
#Step6 - Cosine Similarity
similarity = cosine_similarity(vectors[-1],vectors[:-1])
```

```python
#step7 - Best Match
best_match_index = similarity.argmax()
```

```python
print(data.iloc[best_match_index]["role"])
```

```python
### Taking User Input
resume = input("Enter your resume skills:")
```

```python
# Show All Similarities
for i, score in enumerate(similarity[0]):
    print(data.iloc[i]["role"], "→", score)
```

```python
#Top 2 recommendations
import numpy as np
top_indices = np.argsort(similarity[0])[::-1][:2]
for i in top_indices:
    print("Recommended Role:", data.iloc[i]["role"])
```
